```python
import pandas as pd
import numpy as np
df1=pd.DataFrame({'C1':[11,21,np.nan],'C2':[15,np.nan,np.nan],'C3':[10,20,30]})
df1
```

```python
# Boolen Function
df1.isna()
```

```python
# Inverter to Boolean Function
df1.notna()
```

```python
# Coloumn wise na Values
df1.isna().sum()
```

```python
# NAN Rows are Dropped (Checked Horizontally)
df1=df1.dropna()
df1
```

```python
df1.fillna(value='DEFAULT')
```

```python
# Mean Values of Coloumns are Calculated
df1.fillna(value=df1.mean())
```

```python
# DataSet 1
data={
    "Roll No":[101,102,103,104,105],
    "Name":["Atharva","Bedang","Gayatri","Dhvan","Neha"],
    "Age":[20,"-",21,20,"22"],
    "Marks":[85,90,"-",88,76],
    "Department":["ENTC","CSE","RNA","Mech","AIML"],
    "Admission Date":["01-06-2023","03-06-2023","05-06-2023","07-06-2023","10-06-2023"]
}
df=pd.DataFrame(data)
    # Save as CSV
df.to_csv("data.csv",index=False)
df
```

```python
# Step 2 Replace '-' with NAN to mark missing Values
df.replace("-",np.nan,inplace=True)
df
```

```python
# Convert Coloumns to Numeric where Needed
df["Age"]=pd.to_numeric(df["Age"],errors='coerce')
df["Marks"]=pd.to_numeric(df["Marks"],errors='coerce')
df
```

```python
# Check Missing Values
print("\nMissing Values\n")
print(df.isnull().sum())
```

```python
# Fill missing age with mean
df["Age"].fillna(df["Age"].mean(),inplace=True)
df
```

```python
# Median for Marks
df["Marks"].fillna(df["Marks"].median(),inplace=True)
df
```

```python
print(df.isnull().sum())
```

```python
# Step 5: Fix inconsistent text fromatting
df["Department"]=df["Department"].str.upper()
df
```

```python
# Step 6: Convert Admission Date to standard date format
df["Admission Date"]=pd.to_datetime(df["Admission Date"])
df
```

```python
# Step 7 Save Cleaned data
df.to_csv("cleaned_student_data.csv",index=False)
print("Cleaned data saved to cleaned_student_data.csv")
```
